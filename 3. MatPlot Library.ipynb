##  Univariate - to understand the quality of 1 feature - ex. Histogram, boxplot, pie 


###  Histogram - 

    ** Univariate
    ** takes Continous data Only

```python
plt.figure(figsize=(18,7))
plt.hist(df['mpg'],bins=5)  # code1
plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('MPG Distribution')
```

```python
plt.figure(figsize=(18,7))
plt.hist(x = 'mpg',data = df,bins = 10) # code 2

plt.xlabel('MPG')
plt.ylabel('Frequency')
plt.title('MPG Distribution')
```

```python
plt.figure(figsize=(20,7))
plt.hist(df['wt'],bins=10,facecolor='green',edgecolor='white', orientation='horizontal') # code 3
plt.ylabel('wt')
plt.xlabel('Frequency')
plt.title('Histogram for wt ')
```

```python
plt.figure(figsize=(15,7))
plt.hist(df['gear'],orientation="vertical")

plt.xlabel('Gear')
plt.ylabel('Frequency')
plt.title('Gear Distribution - Discrete Data - Not recommended plotting Hist')
```

### Box Plot

    ** Univariate
    ** for more than 2 columns pass as list 

```python
plt.figure(figsize=(18,7),facecolor='blue')
plt.boxplot(df['mpg'])
plt.xlabel('mpg',size=20)
plt.title('Blox Plot for MPG',size=20)
```

```python
plt.figure(figsize=(18,7),facecolor='blue')
plt.boxplot([df['mpg'],df['wt']])
plt.xlabel('mpg',size=20)
plt.title('Blox Plot for MPG',size=20)
```

```python
plt.figure(figsize=(18,7),facecolor='blue')
plt.boxplot(df)

plt.title('Blox Plot for Data',size=20)
```

```python
# not like x&y in seaborn lib
```

### Violin Plot

    ** Univariate
    ** for more than 1 pass as alist

```python
plt.figure(figsize=(18,7),facecolor='red')
plt.violinplot(df['mpg'])
plt.xlabel('mpg',size=20)
plt.title('Violin Plot for MPG',size=20)
```

```python
plt.figure(figsize=(18,7),facecolor='red')
plt.violinplot(df)
plt.xlabel('X- Axis',size=20)
plt.title('Violin Plot for Entire Data',size=20)
```

### Pie Plot

    ** UniVariate
    ** Data must be Discrete Data 

```python
plt.figure(figsize=(8,8))
plt.pie(x = df['gear'].value_counts(),labels=[1,2,3],autopct='%1.0f%%',explode =[0.02,0.02,0],shadow=True,labeldistance=1.1)
plt.show()
```

```python
plt.figure(figsize= (10,7))
df['gear'].value_counts().plot(kind='pie')
```

```python
plt.figure(figsize=(8,8))
df.value_counts(df['gear']).plot(kind='pie')
```

##  Bivariate - to understand the quality of 2 features - ex. scatter plot, bar Plot

### Bar Chart 

    * Bivariate 
    * Discrete vs Continuous  i.e x axis should be discrete and y - Continuous  

```python
pd.crosstab(df['gear'],df['cyl'])
```

```python
pd.crosstab(df['gear'],df['cyl']).plot(kind='bar')
```

```python
pd.crosstab(df['gear'],df['cyl']).plot(kind='barh')
```

```python
plt.figure(figsize=(18,7))
plt.bar(df['cyl'],df['mpg'])

plt.xlabel('No of Cyl')
plt.ylabel('MPG')
```

### Scatter Plot
    
    ** Bivariate
    ** x & y Must be Continuous Data 

```python
plt.figure(figsize=(18,6))
plt.scatter(df['mpg'],df['wt'])
plt.xlabel('mpg')
plt.ylabel('wt')
```

```python
plt.figure(figsize=(18,6))
plt.scatter(x = df['mpg'], y = df['hp'])
plt.xlabel('mpg')
plt.ylabel('hp')
```

```python
# points
x = [-4, 7, 10]
y = [4, 9, 5]
```

```python
plt.figure(figsize=(18, 6), facecolor='green')

plt.scatter(x, y, s=150, color='blue')  # only points
plt.plot(x, y, linewidth=5, color='red')  # joins the points with line

plt.xlabel('X Axis', size=10)
plt.ylabel('Y Axis', size=10)

plt.xlim(-10, 11)
plt.ylim(0, 10)

plt.title('Line Plot', size=20, color='blue')
plt.suptitle('Sales Comparision Plot', size=20, y=1, color='red')  #if y not mentioned then it will overlap with title

plt.savefig('myplot.jpeg')
```
